## Global Population Estimates
The World Bank's global population estimates --the dataset can be found from this url:-
"https://www.kaggle.com/datasets/theworldbank/global-population-estimates"

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
from matplotlib import pylab
print(pylab.__version__)
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

pd.get_option("display.max_columns")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

### Reading and cleaning the data

```python
data = pd.read_csv('data.csv')
# data.head()

# numeric data
data_numeric = data.select_dtypes(include=[np.number])
numeric_cols = data_numeric.columns.values
print(f'There are {len(numeric_cols)} numeric columns in the dataset','\n')
print(numeric_cols)

# categorical data
data_categorical = data.select_dtypes(exclude=[np.number])
non_numeric_cols = data_categorical.columns.values
print(f'There are {len(non_numeric_cols)} non-numeric columns in the dataset','\n')
print(non_numeric_cols)

for col in numeric_cols:
    data[col] = data[col].fillna(data[col].mean())
    
for non_numeric_col in non_numeric_cols:
    data[non_numeric_col] = data[non_numeric_col].fillna(data[non_numeric_col].value_counts().index[0])
    
data.isnull().sum() #to confirm for missing values(features with zero indicate that it contains no missing value(s))
```

### Sample-point selection based on the regions of the group members (PP22/C626)

```python
Africa_region = data[data['Country Name'].str.contains('Africa')]
Ethiopia_region = data[data['Country Name'].str.contains('Ethiopia')]
Kenya_region = data[data['Country Name'].str.contains('Kenya')]
India_region = data[data['Country Name'].str.contains('India')]
Nigeria_region = data[data['Country Name'].str.contains('Nigeria')]
```

```python
selected_regions = pd.concat([Africa_region,Ethiopia_region,Kenya_region,India_region,Nigeria_region], axis=0)
```

```python
selected_regions.head()
```

```python
selected_regions.shape
```

### Renaming year features

```python
df=pd.DataFrame(selected_regions.iloc[:,4:])
years_rename=list(map(str,range(1960,2051)))
df.columns=years_rename

df_new = pd.DataFrame(selected_regions.iloc[:,:4])
selected_regions = df_new.join(df, how='outer')
```

### World population by selected regions

```python
# plt.figure(figsize=(80,60))
sns.barplot(selected_regions['1960'],selected_regions['Country Name'])
```

```python
# plt.figure(figsize=(80,60))
sns.barplot(selected_regions['2022'],selected_regions['Country Name'])
```

### The above barplots indicate the projection in population of each region, with a few selected years 1960, 2022 and 2050. The horizontal colored lines representing the growth/decline in population while the black lines indicating the uncertainties of the estimates. The country with the highest projection in 1960 is India, with Central Africa Republic being the least, the result was is same in 2022, but is expected to change by/before the year 2050 where sub-saharan African countries will take the lead.

```python
# plt.figure(figsize=(80,40))

data_2=selected_regions.groupby(by="Country Name")["2022"].sum()
res_2=data_2.reset_index()
plt.pie(x="2022",labels="Country Name", data=res_2, autopct="%.2f %%",colors=["r","g","b"])
# plt.xlabel("Sales 2016")
plt.show()
```

### The above is a pie chart representation of the percentages of the regional population in 2022. This is related to the barplots in the previous plots
